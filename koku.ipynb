```python
import zipfile
import os

# Specify the path to your zip file
zip_file_path = 'archive (1).zip'

# Specify the directory where you want to extract the CSV files
extracted_dir = 'path/to/extracted/files/'

# Create the directory if it doesn't exist
os.makedirs(extracted_dir, exist_ok=True)

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# List all the CSV files in the extracted directory
csv_files = [file for file in os.listdir(extracted_dir) if file.lower().endswith('.csv')]

# Display the names of the CSV files
for csv_file in csv_files:
    print(csv_file)
```

```python
import pandas as pd

# Choose one CSV file from the list
selected_csv_file = 'path/to/extracted/files/TCS.csv'

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(selected_csv_file)

# Display basic information about the DataFrame
print("DataFrame Info:")
print(df.info())

# Display the first few rows of the DataFrame
print("\nFirst Few Rows:")
print(df.head())

# Display basic statistics of the numerical columns
print("\nBasic Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(selected_csv_file, parse_dates=['Date'])  # Ensure the 'Date' column is parsed as dates

# Display basic information about the DataFrame
print("DataFrame Info:")
print(df.info())

# Display the first few rows of the DataFrame
print("\nFirst Few Rows:")
print(df.head())

# Plotting examples
plt.figure(figsize=(12, 6))

# Example 1: Line plot for closing prices over time
plt.subplot(2, 1, 1)
plt.plot(df['Date'], df['Close'], label='Closing Price', color='blue')
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()

# Example 2: Bar plot for trading volume
plt.subplot(2, 1, 2)
plt.bar(df['Date'], df['Volume'], label='Volume', color='green')
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()

plt.tight_layout()
plt.show()
```
